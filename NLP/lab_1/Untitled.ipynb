# TOKENIZATION

# Bigram- two consecutive words
# Trigrams- three consecutive words
# Ngram- 'n' number of consecutive words

```python
from nltk.util import bigrams, trigrams, ngrams
```

```python
# use word tokenization 
```

```python
my_string= "Your time is limited, so don't waste it living someone else's life. Don't be trapped by dogma – which is living with the results of other people's thinking."
```

```python
type(my_string)
```

```python
quote_tokenization= word_tokenize(my_string)
```

```python
quote_tokenization_freq= freq_dist.most_common(4)
```
