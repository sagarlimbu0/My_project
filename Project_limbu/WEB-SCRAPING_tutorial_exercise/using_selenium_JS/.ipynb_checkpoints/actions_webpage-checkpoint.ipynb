```python
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
from selenium.common.exceptions import TimeoutException
```

```python
# creating a class with self object
class atleast_select_n(object):
    
    def __init__(self, locator, n):
        self.locator= locator
        self.n= n
        
    # call function
    def __call__(self, browser):
        
        elements= browser.find_elements(*self.locator)
        
        if len(elements) >= self.n:
            return elements
        else:
            return False
```

```python
url='http://www.webscrapingfordatascience.com/complexjavascript/'
```

```python
browser= webdriver.Chrome()
```

```python
browser.get(url)
```

```python
# USE implicit TIME. here we will not use EXPLICIT time frame
browser.implicitly_wait(10)

# here using action to go for INIFINITE scrolling
div_element= browser.find_element_by_class_name('infinite-scroll')
```

```python
# locate the quote by CSS Selector CLASS
quotes_locator= (By.CSS_SELECTOR, '.quote:not(.decode)') #search for quote class and not decode
nr_quotes= 0
```

```python
down_quotes=[]
# while the statement is TRUE
while True:
    #scroll down to the bottom
    browser.execute_script(
        'arguments[0].scrollTop = arguments[0].scrollHeight',
        div_element
    )
    
    # TRY to fetch at least nr_quotes +1 for 3 seconds of lapse time
    try:
        all_quotes= WebDriverWait( browser, 3).until(
            # Following conditions:
            # as parameter takes 2 arguement
            # First parameter: element by class name
            # second param: counts added by each quotes till BOTTOM
            atleast_select_n(quotes_locator, nr_quotes +1)
        )
        
    except TimeoutException as ex:
        # no new quotes found within 3 seconds, assume this is the end
        
        print("..done!!")
        break
        
    # otherwise, update the QUOTE counter
    nr_quotes= len(all_quotes)
    print('.. now seeing', nr_quotes, "quotes")

# print all quotes 
    print(len(all_quotes), 'quotes found \n')

    for quote in all_quotes:
        print(quote.text)
        down_quotes.append(quote.text)
```

```python
input('press Enter to close the automated browser')
browser.quit()
        
```
