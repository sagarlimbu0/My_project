```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
import os
```

```python
df_fg= pd.read_csv('records.csv')
```

```python
df_fg
```

```python
import seaborn as sns; sns.set()
```

```python
sns.lineplot(df_fg['Test'], df_fg['Frog1'], color='r', label= 'Frog1')
sns.scatterplot(df_fg['Test'], df_fg['Frog1'], color='black')

sns.lineplot(df_fg['Test'], df_fg['Frog2'], color='b', label= 'Frog2')
sns.scatterplot(df_fg['Test'], df_fg['Frog2'], color='black')

sns.lineplot(df_fg['Test'], df_fg['Frog3'], color='g', label= 'Frog3')
sns.scatterplot(df_fg['Test'], df_fg['Frog3'], color='black')

sns.lineplot(df_fg['Test'], df_fg['Frog4'], color='y', label= 'Frog4')
sns.scatterplot(df_fg['Test'], df_fg['Frog4'], color='black')
plt.plot(6,5)
plt.ylabel('Jumps (inches)', fontsize=15)
plt.xlabel('Tests', fontsize= 15)
plt.gcf().set_size_inches(9,7)
plt.plot([1,11], [6,6],'--')
plt.plot([1,11], [2,2],'--')
plt.title('Current Baseline', fontsize=16)
plt.ylim(0,8)
plt.legend(loc="upper left")
```

```python
sns.lineplot(df_fg_imprv['Test'], df_fg_imprv['Frog1'], color='r', label= 'Frog1')
sns.scatterplot(df_fg_imprv['Test'], df_fg_imprv['Frog1'], color='black')

sns.lineplot(df_fg_imprv['Test'], df_fg_imprv['Frog2'], color='b', label= 'Frog2')
sns.scatterplot(df_fg_imprv['Test'], df_fg_imprv['Frog2'], color='black')

sns.lineplot(df_fg_imprv['Test'], df_fg_imprv['Frog3'], color='g', label= 'Frog3 (Benchmark)')
sns.scatterplot(df_fg_imprv['Test'], df_fg_imprv['Frog3'], color='black')

sns.lineplot(df_fg_imprv['Test'], df_fg_imprv['Frog4'], color='y', label= 'Frog4')
sns.scatterplot(df_fg_imprv['Test'], df_fg_imprv['Frog4'], color='black')
plt.plot(6,5)
plt.ylabel('Jumps (inches)', fontsize=15)
plt.xlabel('Tests', fontsize= 15)
plt.gcf().set_size_inches(9,7)
plt.ylim(1.2,8)
plt.legend(loc="upper left")
plt.plot([1,11], [6,6],'--')
plt.plot([1,11], [2,2],'--')
plt.title('After Improvements', fontsize=16)
plt.show()
```
